```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('grocery_store_items.csv')
data
```

```python
plt.figure(figsize = (9, 6)) #determine the size of the chart
# To create a bar chart with matplotlib you can use the 'bar' function.
plt.bar(x = data["Grocery Items"], #specify the x axis
        height = data["Units Sold"], #specify the y axis
        color = "midnightblue") # the color for the bars
plt.xticks(rotation = 45, fontsize = 13) # rotate and format the labels for the x-axis
plt.yticks(fontsize = 13) # format the y-axis
plt.title("Grocery items sold", fontsize = 16, fontweight = "bold") #add and format the title for the chart
plt.ylabel("Units Sold", fontsize = 13 ) #add a title for the y-axis
plt.savefig("Grocery items.png") # you can export your chart as a picture
plt.show() # depending on which environment you're using without this line you're chart may not display properly.
```
