# Computer vision ethical implications
https://digitalcommons.spu.edu/cgi/viewcontent.cgi?article=1100&context=honorsprojects

Although computer vision is a promising technology that will surely allow us to harness lots of potential, it also poses some threats that have to be analysed from an ethical perspective. Fundamental issues related to human rights, error rates, and bias have the potential to create a backlash towards the technology, which could limit its benefits as well as harm people in the process.

To avoid these perils, it's clear that tech companies have to be willing to cooperate and be on the same page in terms of wanting to protect consumer privacy and other civil rights. Thus, pressure needs to be applied on tech companies to be more transparent and invite consumers, NGOs, academics and, potentially, other institutions to be a part of the process of development and deployment.

# Image preprocessing
### What did the coder have to do to PROCESS the images in the cat vs. dogs dataset to enable the image classification method to be accurate?

- **Cleaning the dataset:** the author used a fast image browser to check all images in the dogs and cats directories to remove some that were completely meaningless.
- **Label validation:** plotting batches of images with their labels to check that the labels are correctly set.
- **Split the training and validation samples:** If a network is complex enough (if it has enough parameters), it can start overfitting , which means that it is able to learn the specific features of the images in the training dataset. In other words, the network loses its generality, and its ability to classify an arbitrary image as a dog or cat image.
- **Data augmentation:** to reduce loss & increase accuracy in the neural network, new training examples are generated from the ones we already have, in such a way as to increase artificially the size of the training sample.
