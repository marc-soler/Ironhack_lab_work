```python
import pandas as pd
from top100 import *
from tkinter import *
```

```python
# Creating the window
window = Tk()
window.title('Song Recommender')
window.minsize(width=500, height=300)
window.config(padx=20, pady=20)
# Creating a label
my_label = Label(text='Welcome to the Song Recommender!', font=('Arial', 24, 'bold'))
my_label.pack(side='left') #or place(x, y) or grid(column, row)
# Creating a button and making it interactive
def button_clicked():
    print('I was clicked')
button = Button(text='Click me', command=button_clicked)
button.pack()
# Entry field & getting the input
input = Entry(width=10)
input.pack()
print(input.get())
window.mainloop() # this always at the end to keep window running
```
