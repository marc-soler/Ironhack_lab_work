# Bonus Challenge 4: Plot Grouped Data
There are differents ways to go about plotting grouped data with a legend using pandas. Below you will walk through an approach to plot your precip data by season using:

- matplotlib
- and a grouped pandas dataframe

To achieve this plot, you will do the following:

- Create a for loop which groups for your pandas dataframe
- Create a figure as you would normally do using fig, ax = plt.subplots()
- Add a legend to your plot using plt.legend()

```python
fig, (ax) = plt.subplots(1,1, figsize=(15, 10))

colors = {"Winter": "lightgrey",
          "Spring": "green",
          "Summer": "darkgreen",
          "Fall": "brown"}

for label, df in data.groupby("seasons"):
    ax.plot(df.months,
            df.precip,
            "o",
            # The label is the season or the group by object in this case
            label=label,
            color=colors[label])
    
ax.set(title="Challenge 4: Plot Grouped Data\nAverage Precipitation", 
       xlabel="Month", ylabel="Mean precipitation (inches)")
ax.legend()
plt.tight_layout()

# Print each label object which is the group by category - season
for label, df in data.groupby("seasons"):
    print(label)
    
# Print each grouped data
for label, df in data.groupby("seasons"):
    print(df)
            
```
