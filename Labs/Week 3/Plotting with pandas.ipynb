```python
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
sns.set(font_scale=1.5, style="whitegrid")
```

```python
df = pd.read_csv('avg-precip-months-seasons.csv')
```

```python
df.info()
```

# Challenge 1: Plot Precipitation Data

```python
ax = df.plot(color ='tab:orange', marker = 'o')
ax.set_xlabel("Month")
ax.set_ylabel("Mean Precipitation (in inches)")
ax.set_title('Average precipitation by month');
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
months = list(df['months'])
precipitations = list(df['precip'])
ax.bar(months,precipitations, color=(0.2, 0.4, 0.6, 0.6))
ax.set_xlabel("Month")
ax.set_ylabel("Mean Precipitation (in inches)")
ax.set_title('Average precipitation by month')
plt.show()
```

```python
plt.subplot(1, 2, 1)
months = list(df['months'])
precipitations = list(df['precip'])
plt.bar(months,precipitations, color=(0.2, 0.4, 0.6, 0.6))
ax.set_xlabel("Month")
ax.set_ylabel("Mean Precipitation (in inches)")
ax.set_title('Average precipitation by month')
plt.subplot(1, 2, 2)
ax = df.plot(color ='tab:orange', marker = 'o')
ax.set_xlabel("Month")
ax.set_ylabel("Mean Precipitation (in inches)")
ax.set_title('Average precipitation by month')
plt.suptitle('Overall figure title')
plt.tight_layout()
plt.show()
```
